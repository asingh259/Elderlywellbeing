```python
import pandas as pd
import nltk, re, string, collections
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
nltk.download('punkt')

df = pd.read_csv("seniors_advocate_bc_news.csv")
df.head()
```

```python
# Remove all stopwords
from gensim.parsing.preprocessing import remove_stopwords
```

```python
cleandata = {}
cleandata['content'] = []
```

```python
#creating clean dataframe
for i in df['content']:
  clean = remove_stopwords(i)
  cleandata['content'].append(clean)
```

```python
#tokenizing the words
tokenized = [sub.split() for sub in cleandata['content']]
```

```python
#creating bigrams
bigrams =[]
for i in tokenized:
  bigrams.append(ngrams(i, 2))
```

```python
#get the frequency of each bigram in the news
bigramsFreq = []
for i in bigrams:
  bigramsFreq.append(collections.Counter(i))
```

```python
#most popular ngrams in each news
for i in bigramsFreq:
  print(i.most_common(10))
```

```python
#Verifying the frequency of words in the whole dataframe. Nanaimo News Website
corpus = " ".join(cleandata['content'])
```

```python
#Importing Stopwords
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
#Adding the unwanted words into the stopwords list.
stop_words = stopwords.words('english')
stop_words.extend(['RELEASE', 'IMMEDIATE', '(SAFER)', '679-2588', '778', '778-679-2588', 'www.seniorsadvocatebc.ca', 'Ł', 'Œ', 'FOR', 'NEWS','STATEMENT','OPINION-EDITORIAL'])
```

```python
clean_text = []
for word in tokenizedcorpus:
    if word not in stop_words:
      clean_text.append(word)
```

```python
bigramcorpusnonstop = ngrams(clean_text,2)
```

```python
BigramnonstopFreq = collections.Counter(bigramcorpusnonstop)
```

```python
#Verifying the new frequency of words to assure cleaned data.
BigramnonstopFreq.most_common(100)
```

#Creating the cleaned Dataframe for further analysis and modelling.
After removing the unwanted words, it is necessary to filter the text from the original file without those selected words.

```python
final_tokenized = [sub.split() for sub in df['content']]
```

```python
def cleantext(text):
  stop_words = (['RELEASE', 'IMMEDIATE', '(SAFER)', '679-2588', '778', '778-679-2588', 'www.seniorsadvocatebc.ca', 'Ł', 'Œ', 'FOR', 'NEWS','STATEMENT','OPINION-EDITORIAL'])
  bodytext = []
  for word in text:
    if word not in stop_words:
      bodytext.append(word)
  return bodytext
```

```python
cleancolumn = [] 
for i in final_tokenized:
  cleancolumn.append(cleantext(i))
```

```python
#Recreating cleaned text
listofwords = []
for i in cleancolumn:
  listofwords.append(" ".join(i))
```

```python
df['clean'] = listofwords
```

```python
pd.DataFrame(df).to_csv('seniors_advocate_bc_news_cleaned.csv', encoding = 'utf-8-sig')
```
