```python
cleandata = {}
cleandata['body'] = []
```

```python
#creating bigrams
bigrams =[]
for i in tokenized:
  bigrams.append(ngrams(i, 2))
```

```python
#get the frequency of each bigram in the news
bigramsFreq = []
for i in bigrams:
  bigramsFreq.append(collections.Counter(i))
```

```python
#most popular ngrams in each news
for i in bigramsFreq:
  print(i.most_common(10))
```

#Creating the cleaned Dataframe for further analysis and modelling.
After removing the unwanted words, it is necessary to filter the text from the original file without those selected words.

```python
final_tokenized = [sub.split() for sub in df['body']]
```

```python
def cleantext(text):
  stop_words = (['Contacts:' ,
                   #'homes:' ,
                   'facts:' ,
                   'visit:https://workingforyou.gov.bc.ca/' ,
                   'Relationsmedia@bchousing.org' ,
                   'Housing778', '584-2433',
                   'Housingmedia@bchousing.org',
                   #'Media Contacts: Ministry of Municipal Affairs and', remove manually
                   'Housing250', '882-8518',
                   'online:https://www.bchousing.org/homes-for-BC' ,
                   'Housing250-952-0617',
                   'More:', 'Read',
                   'follow.',
                   '952-0617',
                   'Relations778', '678-1572','584-2433', '698-9176', '584-1255',
                   '698-9176',
                   'Housing604', '439-8571', '456-8917',
                   'Quick', 'Facts:'
                   'Corporation604', '787-1787',
                   'HousingMedia',
                   'Columbians:bcbudget.gov.bc.ca/2018/homesbc/2018_Homes_For_BC.pdf',
                   'visit:https://news.gov.bc.ca/factsheets/bc-government-addressing-housing-affordability-challenges'
                    ])
  bodytext = []
  for word in text:
    if word not in stop_words:
      bodytext.append(word)
  return bodytext
```

```python
cleancolumn = [] 
for i in final_tokenized:
  cleancolumn.append(cleantext(i))
```

```python
#Recreating cleaned text
listofwords = []
for i in cleancolumn:
  listofwords.append(" ".join(i))
```

```python
df['clean'] = listofwords
```
