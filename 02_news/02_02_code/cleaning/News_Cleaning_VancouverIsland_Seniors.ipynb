```python
import pandas as pd
import nltk, re, string, collections
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
nltk.download('punkt')

df = pd.read_csv("seniors_vancouverisland_seniors.csv")
df.head()
```

```python
# Remove all stopwords
from gensim.parsing.preprocessing import remove_stopwords
```

```python
cleandata = {}
cleandata['body'] = []
```

```python
#creating clean dataframe
for i in df['body']:
  clean = remove_stopwords(i)
  cleandata['body'].append(clean)
```

```python
#tokenizing the words
tokenized = [sub.split() for sub in cleandata['body']]
```

```python
#creating bigrams
bigrams =[]
for i in tokenized:
  bigrams.append(ngrams(i, 2))
```

```python
#get the frequency of each bigram in the news
bigramsFreq = []
for i in bigrams:
  bigramsFreq.append(collections.Counter(i))
```

```python
#most popular ngrams in each news
for i in bigramsFreq:
  print(i.most_common(10))
```

#Creating the cleaned Dataframe for further analysis and modelling.
After removing the unwanted words, it is necessary to filter the text from the original file without those selected words.

```python
final_tokenized = [sub.split() for sub in df['body']]
```

```python
def cleantext(text):
  stop_words = (['onFacebookand','onTwitter.Want','onTwitter.BC','PressLike',"googletag.display('div-gpt-3');gptRefreshSlots.push(3);","googletag.display('div-gpt-4');gptRefreshSlots.push(4);","});googletag.cmd.push(function()","googletag.display('div-gpt-56');gptRefreshSlots.push(56);","googletag.display('div-gpt-5');gptRefreshSlots.push(5);", "googletag.display('div-gpt-57');gptRefreshSlots.push(57);","{","PressRELATED:",
                   'pleaseclick','donationhere.','donationhere','onTwitter.','onTwitter','onTwitterandInstagram,and','onFacebook.'])
  bodytext = []
  for word in text:
    if word not in stop_words:
      bodytext.append(word)
  return bodytext
```

```python
cleancolumn = [] 
for i in final_tokenized:
  cleancolumn.append(cleantext(i))
```

```python
#Recreating cleaned text
listofwords = []
for i in cleancolumn:
  listofwords.append(" ".join(i))
```

```python
df['clean'] = listofwords
```

```python
pd.DataFrame(df).to_csv('seniors_vancouverisland_seniors_cleaned.csv', encoding = 'utf-8-sig')
```
