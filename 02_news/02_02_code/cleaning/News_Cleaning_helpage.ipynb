```python
import pandas as pd
import nltk, re, string, collections
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
#nltk.download('punkt')
```

```python
df = pd.read_csv("helpagecanada_news_manualedited.csv")
df.head()
```

```python
df['dates'] = pd.to_datetime(df['dates'])
df['dates'] = pd.to_datetime(df["dates"].dt.strftime('%d %B %Y'))
df.head()
```

```python
len(df)
```

```python
# Remove all stopwords
from gensim.parsing.preprocessing import remove_stopwords
```

```python
cleandata = {}
cleandata['body'] = []
```

```python
#creating clean dataframe
for i in df['body']:
  clean = remove_stopwords(i)
  cleandata['body'].append(clean)
```

```python
#tokenizing the words
tokenized = [sub.split() for sub in cleandata['body']]
```

```python
#creating bigrams
bigrams =[]
for i in tokenized:
  bigrams.append(ngrams(i, 2))
```

```python
#get the frequency of each bigram in the news
bigramsFreq = []
for i in bigrams:
  bigramsFreq.append(collections.Counter(i))
```

```python
#most popular ngrams in each news
for i in bigramsFreq:
  print(i.most_common(10))
```

```python
#Verifying the frequency of words in the whole dataframe. Nanaimo News Website
corpus = " ".join(cleandata['body'])
```

```python
bigramcorpus = ngrams(tokenizedcorpus,2)
```

Based on above words found the body column is carefully evaluated and found exact long related stop words. after removing stop words we will remove strongs containing @ and https:// etc..

```python
#Importing Stopwords
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
#Adding the unwanted words into the stopwords list.
stop_words = stopwords.words('english')
stop_words.extend(['June', '15,','25,', '2021','1975.','205-1300', 'ON','K1Z','7L2', '613-232-0727', 'Toll', 'Free:1-800-648-1111',
'info@helpagecanada.ca','Privacy', 'Policy', 'Contact',  '©', 'HelpAge', 'Canada', 'Number:', '118', '955', '921', 'RR0001'])
```

```python
clean_text = []
for word in tokenizedcorpus:
    if word not in stop_words:
      clean_text.append(word)
```

```python
bigramcorpusnonstop = ngrams(clean_text,2)
```

```python
BigramnonstopFreq = collections.Counter(bigramcorpusnonstop)
```

```python
#Verifying the new frequency of words to assure cleaned data.
BigramnonstopFreq.most_common(500)
```

#Creating the cleaned Dataframe for further analysis and modelling.
After removing the unwanted words, it is necessary to filter the text from the original file without those selected words.

```python
final_tokenized = [sub.split() for sub in df['body']]
```

```python
def cleantext(text):
  stop_words = (['June', '15,','25,', '2021','1975.','205-1300', 'ON','K1Z','7L2', '613-232-0727', 'Toll', 'Free:1-800-648-1111',
'info@helpagecanada.ca','Privacy', 'Policy', 'Contact',  '©', 'HelpAge', 'Canada', 'Number:', '118', '955', '921', 'RR0001'])
  bodytext = []
  for word in text:
    if word not in stop_words:
      bodytext.append(word)
  return bodytext
```

```python
cleancolumn = [] 
for i in final_tokenized:
  cleancolumn.append(cleantext(i))
```

```python
#Recreating cleaned text
listofwords = []
for i in cleancolumn:
  listofwords.append(" ".join(i))
```

```python
df['clean'] = listofwords
```

```python
pd.DataFrame(df).to_csv('helpage_news_cleaned.csv', encoding = 'utf-8-sig')
```
